```python
#function to plot random images from the training set
def plot_sample_images(data, n):
    '''
    data = numpy array show shape (batch_size, h, w, 3) for rgb images or shape (batch_size, h, w, 1) for gray scale images
    n = number of random images to plot
    '''
    total_num_images = len(data)
    
    #initialize plot
    fig, ax = plt.subplots(1,n)
    for i in range(0, n):
        rand_index = np.random.choice(np.arange(0,total_num_images))
        ax[i].imshow(data[rand_index])
        ax[i].set_title('Image' + " " + str(i+1) + ": " + label_dict[y_train[rand_index]])
```

## Model 1: Pre-Trained VGG16 + Fully Connected Layers

```python
#load layers and pre-trained model

#load keras model to initialize model
from keras import Model

#load layers required for model and regularization
from keras.layers import Input
from keras.layers import RandomRotation
from keras.layers import RandomZoom
from keras.layers import RandomContrast
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import Dense
from keras.regularizers import l1_l2

#load pre-trained model
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input

#from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import plot_model

#loead pre-trained model excluding top layer and freeze weights
pre_trained_model = VGG16(weights='imagenet', include_top=False, input_shape=(125,94,3))
for layer in pre_trained_model.layers:
    layer.trainable = False
```

```python
#define model
    
#Multiple Image Inputs
input_img = Input(shape=(125,94,3), name='Input_Layer_Img')

#Augmentation Layers
input_img_augmented = RandomRotation(factor=0.2, name='Augmentation_Layer_Image')(input_img)
input_img_augmented = RandomZoom(height_factor=(-0.2, 0.2), width_factor=(-0.2, 0.2))(input_img_augmented)
input_img_augmented = RandomContrast(factor=0.9)(input_img_augmented)

#pre-proecss layer for ResNet50                                    
input_pre_processed = preprocess_input(input_img_augmented)

#get features embeddings from pre-trained 
embeddings_img = pre_trained_model(input_pre_processed)

#flattem embeddings
embeddings_flat = Flatten(name='Flatten_Layer_img_embedding')(embeddings_img)

#dropout layer
dropout_01 = Dropout(0.5, name='Dropout_layer_01')(embeddings_flat)

#dense layer
regularizer = l1_l2(l1=0.1, l2=0.1)
dense = Dense(units=100, activation='relu', name='Dense_Hidden_Layer', kernel_regularizer=regularizer)(dropout_01)

#dropout layer
dropout_02 = Dropout(0.5, name='Dropout_layer_02')(dense)

#dense output layer
regularizer = l1_l2(l1=0.1, l2=0.1)
output = Dense(units=len(np.unique(y_train)), activation='sigmoid', name='Dense_Output_Layer', kernel_regularizer=regularizer)(dropout_02)

#initialize model
model_01 = Model(inputs=input_img, outputs=output)
```

```python
#show model summary
model_01.summary()
```

```python
#plot model
plot_model(model = model_01, to_file='model_01.png')
```

```python
# compile the model (should be done *after* setting layers to non-trainable)

#load optimization function
from tensorflow.keras.optimizers import Adam
optimization_function = Adam(learning_rate=0.001,
                             beta_1=0.9,
                             beta_2=0.999,
                             epsilon=1e-07,
                             amsgrad=False,
                             name="Adam")

#load loss function
from keras.losses import SparseCategoricalCrossentropy #as labels not one-hot encoded
loss_function = SparseCategoricalCrossentropy(from_logits=False)

#compile model with definined optimization and loss function
model_01.compile(optimizer=optimization_function, loss=loss_function, metrics=['accuracy'])
```

```python
#train model
from keras.callbacks import EarlyStopping
callback_early_stopping = EarlyStopping(patience=50)

history_01 = model_01.fit(
                    x=X_train,
                    y=y_train,
                    batch_size=128,
                    epochs=500,
                    validation_split=0.1,
                    shuffle=True,
                    workers=-1,
                    use_multiprocessing=True,
                    callbacks = callback_early_stopping
                    )
```

```python
def plot_model_training(training_history, metric):
    '''
    Function plots the desiered metric over training epocs on both the training and validation set
    
    Parameters
    training_history = history of training data. output from tensorflow from model fitting
    metic = desired metric to plot over epocs. 'loss' for training and validation loss. 'accuracy' for training and validation accuracy
    
    '''
  
    fig, ax = plt.subplots(figsize=(10,6))
    ax.plot(training_history.history[metric], label='training_'+metric)
    ax.plot(training_history.history['val_' + metric], ls ='dashed', label='validation_'+metric)
    ax.set_title('Train and validation ' + metric + ' per training epoch')
    ax.set_xlabel('epoch')
    ax.set_ylabel(metric)
    ax.legend()
```

```python
plot_model_training(training_history=history_01, metric='loss')
```

```python
plot_model_training(training_history=history_01, metric='accuracy')
```

```python
#model performance append
model_performance.loc[0,'Model'] = 'Model 1: Pre-Trained VGG16 + Fully Connected Layers'
model_performance.loc[0,'Validation_Loss'] = history_01.history['val_loss'][-1]
model_performance.loc[0,'Validation_Accuracy'] = history_01.history['val_accuracy'][-1]
```

## Model 6: Pre-Trained Xception + Convolutional Layers + Fully Connected Layers

```python
#load layers and pre-trained model

#load keras model to initialize model
from keras import Model

#load layers required for model and regularization
from keras.layers import Input
from keras.layers import RandomRotation
from keras.layers import RandomZoom
from keras.layers import RandomContrast
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.regularizers import l1_l2

#load pre-trained model
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.applications.xception import preprocess_input

#from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import plot_model

#loead pre-trained model excluding top layer and freeze weights
pre_trained_model = Xception(weights='imagenet', include_top=False, input_shape=(125,94,3))
for layer in pre_trained_model.layers:
    layer.trainable = False
```

```python
#define model
    
#Multiple Image Inputs
input_img = Input(shape=(125,94,3), name='Input_Layer_Img')

#Augmentation Layers
input_img_augmented = RandomRotation(factor=0.2, name='Augmentation_Layer_Image')(input_img)
input_img_augmented = RandomZoom(height_factor=(-0.2, 0.2), width_factor=(-0.2, 0.2))(input_img_augmented)
input_img_augmented = RandomContrast(factor=0.9)(input_img_augmented)

#pre-proecss layer for ResNet50                                    
input_pre_processed = preprocess_input(input_img_augmented)

#get features embeddings from pre-trained 
embeddings_img = pre_trained_model(input_pre_processed)

#conolutional layers
conv_01 = Conv2D(filters=64, kernel_size=3, activation ="relu", padding ="same", name='Conv_Layer_01')(embeddings_img)
max_pool_01 = MaxPooling2D(pool_size=2, name='Max_Pool_Layer_01')(conv_01)
conv_02 = Conv2D(filters=128, kernel_size=3, activation ="relu", padding ="same", name='Conv_Layer_02')(max_pool_01)
conv_03 = Conv2D(filters=256, kernel_size=3, activation ="relu", padding ="same", name='Conv_Layer_03')(conv_02)

#flatten layer
flatten = Flatten(name='Flatten_Layer_01')(conv_03)

#dropout layer
dropout_01 = Dropout(0.5, name='Dropout_Layer_01')(flatten)

#dense layer
regularizer = l1_l2(l1=0.1, l2=0.1)
dense = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_01', kernel_regularizer=regularizer)(dropout_01)

#dropout layer
dropout_02 = Dropout(0.5, name='Dropout_Layer_02')(dense)

#dense output layer
regularizer = l1_l2(l1=0.1, l2=0.1)
output = Dense(units=len(np.unique(y_train)), activation='sigmoid', name='Dense_Output_Layer', kernel_regularizer=regularizer)(dropout_02)

#total model
model_06 = Model(inputs=input_img, outputs=output)
```
